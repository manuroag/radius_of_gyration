### Estructura Hélice Alfa de las proteínas
Cuando la cadena polipeptídica se enrolla en espiral sobre sí misma debido a 
los giros producidos en torno al carbono alfa de cada aminoácido se adopta una 
conformación denominada héliceα. Esta estructura es periódica y en ella cada enlace 
peptídico puede establecer dos puentes de hidrógeno.



Así, nuestro approach será convertir la estructura compleja de la proteína (ver inciso A de la imagen anterior) a una más simple (Inciso B de la imagen) encontrando el centro de masa de cada aminoácido que compone a la proteína, que como ya dijimos son los monómeros unidad de nuestro polímero (proteína), para luego calcular el correspondiente radio de giro, dado por la ecuación:

$${R_g}^2 = \sum_{i=1}^{N} (R_i - R_{CM} )^2 / N + \frac{3}{5} R^2 $$

Para posteriormente graficar el radio de giro $Rg$ de cada proteína vs el número de aminoácidos $N$ (longitud de la cadena).

## Código
Usaremos una librería llamada "Biopython" que nos facilitará muchísimo la lectura de nuestros archivos .PDB extraídos del Protein Data Bank.

```python
from Bio import PDB
```

```python
def get_protein(ruta):
    # We get the protein structure and the global center of mass
    parser = PDB.PDBParser(QUIET=True)
    # We get the first model [0] (Crystal structures generally have only one model (with id 0), while NMR files usually have several models)
    protein = parser.get_structure('test', ruta)[0]
    CM_protein = protein.center_of_mass()
    return protein,CM_protein
```

Ahora debemos limpiar un poquito los datos, pues no todos los residuos que se nos dan son aminoácidos, podemos encontrar moléculas de agua por ejemplo. Así imponemos la condición para ignorar todo lo que no sea un aminoácido con ayuda de la función:

PDB.is_aa(residue)

### Funciones

```python
def centers_of_mass(protein):
    # We get all amino acids' centers of mass, ignoring molecules that are not amino acids (eg. water molecules)  
    CM_residues=[]
    N=0 # Length of the chain (Number of amino acids in the protein)
    for chain in protein:
        for residue in chain:
            if PDB.is_aa(residue):
                CM_residues.append(residue.center_of_mass())
                N+=1
    return CM_residues,N
```

```python
def radius_of_gyration(CM_protein,CM_residues,N):
    # We calculate the Radius of Gyration
    suma=0
    for r in CM_residues:
        vector=[r[0]-CM_protein[0], r[1]-CM_protein[1], r[2]-CM_protein[2]]
        magnitud = vector[0]**2 + vector[1]**2 + vector[2]**2
        suma+=magnitud
    suma=suma/N
    suma+=(3/5)*((1.5)**2)
    return suma**(1/2)
```

Una vez definidas estas funciones, ya podemos extraer nuestros datos de proteínas y leerlos para extraer de cada uno el radio de giro y su longitud.

### Main

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
rgyration=[]
polymerization=[]

for i in range(1,38):
    # We get alpha-helix type proteins and their center of mass
    ruta = os.path.join('..', 'data/alpha', "A"+str(i)+".pdb")
    protein,CM_protein = get_protein(ruta)
    # We get the center of mass per residue per protein and polymerization (number of aminoacids N)
    CM_residue,N = centers_of_mass(protein)
    polymerization.append(N)
    # We calculate the radius of gyration of the protein
    Rg=radius_of_gyration(CM_protein,CM_residue,N)
    rgyration.append(Rg)
    
#print(f'Radius of Gyration: {rgyration}')
#print(f'\nLenght of chains: {polymerization}')
```

```python
data = np.column_stack((polymerization,rgyration))
df = pd.DataFrame(data, columns=["N","Rg"])
df.head()
```

Graficamos nuestros datos experimentales

```python
plt.scatter(df["N"], df["Rg"], color="m", marker='*', s=20, label="Proteínas")
plt.xlabel("Número de aminoácidos N")
plt.ylabel("Radio de giro (Rg)")
plt.title("Proteínas alfa hélice")
plt.grid(ls="--", color="#dadada")

plt.legend(loc=2)

plt.show()
```

### Ajuste de nuestros datos

```python
lnData = np.log(data)
lndf = pd.DataFrame(lnData, columns=["lnN","lnRg"])
lndf.head()
```

```python
plt.scatter(lndf['lnN'],lndf['lnRg'], color="m", s=20, marker='*', label="Proteínas")
plt.xlabel("ln(N)")
plt.ylabel("ln(Rg)")
plt.title("Gráfica logarítmica")
plt.grid(ls="--", color="#dadada")

# Ajuste Lineal
x = np.linspace(np.min(lndf['lnN'])-0.1, np.max(lndf['lnN'])+0.1)
coeficiente = np.polyfit(lndf['lnN'],lndf['lnRg'],1)

plt.plot(x, coeficiente[0]*x + coeficiente[1], label=f"$ln(Rg)={coeficiente[0]:.3f}*ln(N) + {coeficiente[1]:.3f}$")

plt.legend()

plt.show()
```

```python
print(f'v = {coeficiente[0]}')
print(f'b = {coeficiente[1]}')
```

### Ajuste detallado

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula='lnRg ~ lnN', data=lndf).fit()

# Print the full summary table
print(model.summary())

# Residual Standard Error (RSE) of the model
rse = np.sqrt(model.scale)
print(f"\nResidual Standard Error (RSE): {rse}")
```

Haciendo un ajuste lineal con _statsmodels_ library obtenemos el exponente así como su incertidumbren, obteniendo que nuestros datos se comportan como:

$$Rg \sim N^{\nu}$$

donde

$$\nu = 0.418 \pm 0.018$$

Graficamos el ajuste:

$$Rg = 1.83 N^{0.418}$$

con coeficiente de correlación:

$$R^2 = 0.94$$

```python
plt.scatter(df["N"], df["Rg"], color="m", marker='*', s=20, label="Proteínas")
plt.xlabel("Número de aminoácidos (N)")
plt.ylabel("Radio de giro (Rg)")
plt.title("Proteínas alfa hélice")
plt.grid(ls="--", color="#dadada")

x = np.linspace(np.min(df['N']), np.max(df['N']))
rg2 = np.exp(coeficiente[1])*((x)**(coeficiente[0]))
plt.plot(x,rg2,label="$R_g \sim N^{0.418 \pm 0.018}$")

plt.legend(loc=2)
plt.show()
```

Así, podemos concluir que según este estudio estadístico, el exponente de escalamiento del radio de giro para las proteínas tipo alfa es de

$$ \nu = 0.418 \pm 0.018 $$
